Ejemplos de Cálculo diferencil e integral.

x = Symbol("x")

```python
diff(sin(x)*exp(x),x)
```

```python
diff(x**2*exp(x),x)
```

```python
integrate(x**2*exp(x),x)
```

```python
integrate(x**200*exp(x),x)
```

```python
lim(sinn(x)/x,x,0)
```

```python
lim(sin(x)/x,x,0)
```

```python
limit(sinn(x)/x,x,0)
```

```python
limit(sin(x)/x,x,0)
```

```python
integrate(x**2,(x,0,1))
```

```python
integrate(sin(x**2),(x,0,1))
```

```python
integrate(sin(x**2),(x,-oo,oo))
```

```python
solve(x**2-2, x)
```

```python
solve(x**2+1,x)
```

```python
a,b,c=symbols("a b c")
```

```python
solve(a*x**2+b*x+c, x)
```

```python
d=Symbols("d")
```

```python
d=Symbol("d")
```

```python
solve(a*x**3+b*x**2+c*x+d, x)
```
